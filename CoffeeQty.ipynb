```python
q1=df1.quantile(0.25)
q3=df1.quantile(0.75)
q1
```

```python
q3
```

```python
iqr=q3-q1
iqr
```

```python
b=(df1<(q1-1.5*iqr))|(df1>(q3+1.5*iqr))
b
```
