```python
x=z.drop(columns=['purchase_amount'])
y=z['purchase_amount']
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.tree import DecisionTreeRegressor
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
decision_tree_regressor=DecisionTreeRegressor(max_depth=2)
```

```python
decision_tree_regressor.fit(x_train,y_train)
```

```python
from sklearn import tree
plt.figure(figsize=(10,10))
tree.plot_tree(decision_tree_regressor,filled=True)
```

```python
pre=decision_tree_regressor.predict(x_test)
print(f"The error rate of the model is {mean_squared_error(y_test,pre)} and the performance is {r2_score(y_test,pre)}")
```

```python
tree_regressor=DecisionTreeRegressor()
```

```python
parameter={    'criterion': ['mse', 'friedman_mse', 'mae'],
    'splitter': ['best', 'random'],
    'max_depth': [1, 2, 3, 4, 5],
    'max_features': ['auto', 'sqrt', 'log2']

}
```

```python
model=GridSearchCV(tree_regressor,parameter,cv=5,scoring='neg_mean_squared_error')
```

```python
model.fit(x_train,y_train)
```

```python
pr=model.predict(x_test)
print(f"The error rate of the model is {mean_squared_error(y_test,pr)} and the performance is {r2_score(y_test,pr)}")
```
